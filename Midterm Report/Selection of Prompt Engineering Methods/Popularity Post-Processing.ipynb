```python
# Load semantic_scholar_df.csv

import pandas as pd

semantic_scholar_df = pd.read_csv('semantic_scholar_df.csv')

semantic_scholar_df
```

```python
# Append on extra_semantic_scholar_df.csv and deduplicate

extra_semantic_scholar_df = pd.read_csv('extra_semantic_scholar_df.csv')

semantic_scholar_df = pd.concat([semantic_scholar_df, extra_semantic_scholar_df], ignore_index=True)

semantic_scholar_df = semantic_scholar_df.drop_duplicates()

semantic_scholar_df
```

```python
# Ensure no duplicates on 'semantic scholar title'

no_duplicates_df = semantic_scholar_df.drop_duplicates(subset=['semantic scholar title'])

print(len(no_duplicates_df))
```

```python
# Print any duplicates on 'semantic scholar title'

duplicates_df = semantic_scholar_df[semantic_scholar_df.duplicated(subset=['semantic scholar title'], keep=False)]

# Sort by 'semantic scholar title' to make it easier to read
duplicates_df = duplicates_df.sort_values(by=['semantic scholar title'])

duplicates_df
```

```python
# Actually probably OK to leave in duplicates for now, at least until after the join...
```

```python
# Load General Text-to-Text Prompt Engineering Methods.xlsx

general_methods_df = pd.read_excel('General Text-to-Text Prompt Engineering Methods.xlsx')

general_methods_df
```

```python
# Drop duplicated rows in semantic_scholar_df
semantic_scholar_df = semantic_scholar_df.drop_duplicates()

# Trim whitespace in 'paper title' column
semantic_scholar_df['paper title'] = semantic_scholar_df['paper title'].str.strip()

# In the general_methods_df, trim whitespace in 'Name of Original Paper' column
general_methods_df['Name of Original Paper'] = general_methods_df['Name of Original Paper'].str.strip()
```

```python
# Merge on 'Method' from general_methods_df
# Merge key is 'paper title' in semantic_scholar_df and 'Name of Original Paper' in general_methods_df
merged_df = pd.merge(semantic_scholar_df, general_methods_df, how='outer', left_on='paper title', right_on='Name of Original Paper')
merged_df
```

```python
# Split off items missing 'semantic scholar title'
missing_semantic_scholar_title_df = merged_df[merged_df['semantic scholar title'].isnull()]
missing_semantic_scholar_title_df

# Remove from merged_df
merged_df = merged_df.dropna(subset=['semantic scholar title'])
merged_df
```

```python
# For any duplicates on 'semantic scholar title' - if there is a row with a value in 'Method' column, keep that row and drop others
merged_df = merged_df.sort_values(by=['semantic scholar title', 'Method'], ascending=False, na_position='last')
merged_df = merged_df.drop_duplicates(subset=['semantic scholar title'], keep='first')

merged_df
```

```python
# Sort by citations_per_day descending
merged_df = merged_df.sort_values(by=['citations_per_day'], ascending=False)
merged_df
```

```python
# Add back on missing_semantic_scholar_title_df
merged_df = pd.concat([merged_df, missing_semantic_scholar_title_df], ignore_index=True)
```

```python
# Output to Excel
merged_df.to_excel('merged_df.xlsx', index=False)
```
